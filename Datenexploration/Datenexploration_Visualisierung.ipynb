# Merging CSV files to one CSV file

```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from math import pi
from matplotlib.pyplot import fill_between
from scipy.stats import skew, kurtosis

# Specify the path to CSV files
csv_files_path = '../Datenbereinigung/*.csv'

# Use glob to get a list of all CSV files in the directory
csv_files = glob.glob(csv_files_path)

# Initialize an empty list to store individual dataframes
dfs = []

# Loop through each CSV file and read it into a dataframe
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    dfs.append(df)

# Concatenate all dataframes vertically
combined_df = pd.concat(dfs, ignore_index=True)

# Now, combined_df contains the vertically stacked data from all CSV files

# Save the combined dataframe as a new CSV file
combined_df.to_csv('combined_data.csv', index=False)
```

```python
# Selecting only numeric columns for correlation matrix
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Plotting the correlation matrix using seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Top Speed', data=df, hue='Marke')
plt.title('Scatter plot of Price vs. Top Speed')
plt.xlabel('Price')
plt.ylabel('Top Speed')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Marke', y='Price', data=df)
plt.title('Average Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Marke', y='Price', data=df)
plt.title('Price Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution plot for Engine Size
plt.figure(figsize=(12, 8))
sns.histplot(df['Engine Size'], bins=20, kde=True, color='purple', edgecolor='black')
plt.title('Distribution of Engine Size', fontsize=16)
plt.xlabel('Engine Size', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.xlim(0, 5500)  # Set x-axis limits to show values less than 7000

plt.tight_layout()
plt.show()
```
